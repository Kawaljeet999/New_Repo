```python
for column in df.columns:  ## Culprit
    if column == "Churn": # Skips Churn column because somehow it make Churn label remove all 1's Values
        continue
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR # (Here we make the lower bound less than Q1)
    upper_bound = Q3 + 1.5*IQR # (Here we make the upper bound greater than Q3)
    strat_train_set = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

## PCA

```python
# We do Pca to spread data and its variance to its features
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3) # n_components=2: This specifies that we want to reduce the data to 2 principal components (i.e., we want to transform the data into 2 new dimensions). PCA will find the two directions (axes) that explain the most variance in the data, and those will be our new features. Mean 3 features
pca_component = pca.fit_transform(df_data) # fit(): This calculates the principal components and learns how to best transform the data. It looks for the directions in the data with the most variance.
                                                         # transform(): This applies the learned transformation to the original data, projecting it onto the principal components.
pca_df = pd.DataFrame(data=pca_component,columns=['PC1','PC2','PC3']) # principal_components will hold the transformed data, now reduced to 2 dimensions. This will be a new dataset where each data point is represented by its position on the two principal components.
```

```python
pca_df_transformed = scaler.fit_transform(pca_df)
# Convert the numpy array to a pandas DataFrame and assign custom column names # the 0,1,2 feature labels due to it is problem by default with scaler in pca
pca_df_transformed = pd.DataFrame(pca_df_transformed, columns=['PC1', 'PC2', 'PC3']) # Convert column name in PC1,PC2 and PC3 from 0,1,2
pca_df_transformed.shape
```

```python
pca_df_transformed.head()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming pca_df is the DataFrame containing the principal components
fig = plt.figure(figsize=(10, 8))

# 3D scatter plot
ax = fig.add_subplot(111, projection='3d') # In essence, 111 is equivalent to saying "create one plot in a single row and column layout, and this is the first plot."
ax.scatter(pca_df['PC1'], pca_df['PC2'], pca_df['PC3'])

# Adding labels and title
ax.set_title('PCA of the Churn Dataset')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

plt.show()
```

## Model Selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# model = LogisticRegression() # 85 % accurate
# model = DecisionTreeClassifier() # 99 % accurate
model = RandomForestClassifier(n_estimators = 100,random_state = 42) # 99 % accurate
# model = GaussianNB() # 85 % accurate
# model = KNeighborsClassifier() # 86 % accurate # default value of k = 5 #
                                                 # Curse of Dimensionality: As the number of features grows, data points become equidistant, making it harder for KNN to distinguish between them.
                                                 # here when we have mutiple features then KNN uses distance while adjacent values in featiures are substracted
# model = SVC(kernel = 'linear') # 85 % accurate
model.fit(pca_df_transformed,df_labels) # Here labels not transformed
```

## Cross validation

```python
from sklearn.model_selection import cross_val_score # Do when the model Overfits

cv_scores = cross_val_score(model,pca_df_transformed,df_labels,cv = 5, scoring = "accuracy") # The scoring parameter defines the metric that will be used to evaluate the model’s performance.


print("Cross-Validation Accuracy Scores:", cv_scores)
print(f"Mean Accuracy: {cv_scores.mean():.4f}")
print(f"Standard Deviation of Accuracy: {cv_scores.std():.4f}")
```

## Fine tuning

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    "n_estimators":[50,100,150],
    "max_depth":[None,10,20] ,# Here if depth is high it overfits and low depth mean underfit
    "min_samples_split":[2,5,10], # min_samples_split is if i split a node and get 2 child nodes mean we have min_smaples_split is 2
                                 # Here 2 may overfitts data due to it make high depth of splitting
                                 # here 10 may underfit data not get good depth
                                 # Higher depth leads to Overfitting and lower leads to underfitting
    "min_samples_leaf":[1,2,4],  # it mean minimum n tuples must pass/ reach to leaf node to achieve a leaf prediction
                                #If min_samples_leaf is too small (e.g., 1), the model may create many splits, leading to overfitting.
                                # If it's large, the tree may be more conservative, resulting in underfitting.
    "max_features":['sqrt','log2'],  #  By selecting a subset of features at each split, the algorithm prevents all trees from being identical (especially in Random Forests).
                                #This increases the diversity among the trees, reducing the risk of overfitting.
                                # If the dataset has 16 features, then sqrt(16)  = 4  so 4 features are considered at each split.
                                # If the dataset has 16 features, then log2(16)=4, so 4 features are considered.
                                # This will help to make every decisio tree with random features
     "bootstrap":[True,False]   # Random samples are drawn with replacement from the dataset.
                                # With bootstrapping (bootstrap=True), a random sample of size 100 is drawn with replacement, meaning some samples might appear multiple times while others might not appear at all.
                                # Without bootstrapping (bootstrap=False), each tree is trained on the entire dataset, without sampling.
}
```

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=model,param_grid=param_grid,cv=5,scoring='accuracy',n_jobs=-1,verbose=1)

# Fitting
grid_search.fit(pca_df_transformed,df_labels) ## here 324 is parametr combination and 1620 evaluation

# For more detail we go in Grid Search Test file for paramerts and workflow
```

```python
# Print best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Score:", grid_search.best_score_)
```
